### 1. Why are functions advantageous to have in your programs?

Advantages of Functions:

Encapsulate reusable code.
Enhance code readability.
Modularize programs into smaller, manageable chunks.
Promote code reusability and maintainability.

### 2. When does the code in a function run: when it&#39;s specified or when it&#39;s called?

When Does Code in a Function Run:

The code in a function runs when the function is called or invoked.

### 3. What statement creates a function?

Statement to Create a Function:

The def statement is used to create a function in Python.


### 4. What is the difference between a function and a function call?

Difference Between a Function and a Function Call:

A function is a block of code that performs a specific task when called.
A function call is the act of invoking or executing a function.

### 5. How many global scopes are there in a Python program? How many local scopes?

Global and Local Scopes:

There's one global scope in a Python program.
Local scopes are created whenever a function is called, and they cease to exist once the function completes execution.


### 6. What happens to variables in a local scope when the function call returns?

Variables in Local Scope After Function Call:

Local scope variables are destroyed once the function call returns, making them inaccessible outside the function.


### 7. What is the concept of a return value? Is it possible to have a return value in an expression?

Return Value:

A return value is the value that a function returns after its execution.
Yes, it's possible to use a return value in an expression.


### 8. If a function does not have a return statement, what is the return value of a call to that function?

Function Without a Return Statement:

If a function lacks a return statement, its return value is None.


### 9. How do you make a function variable refer to the global variable?

Referencing Global Variables in Functions:

Use the global keyword to modify a global variable from within a function.

### 13. What can you do to save a programme from crashing if it encounters an error?

Handling Errors to Prevent Crashing:

Use try and except blocks to catch and handle exceptions gracefully, preventing the program from crashing.
